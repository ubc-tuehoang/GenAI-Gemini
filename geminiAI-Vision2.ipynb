```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
```

```python
def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
genai.configure(api_key="<YOUR API TOKEN>")
```

```python
import PIL.Image

img = PIL.Image.open('17111219496980102-a.png')
img
```

```python
model = genai.GenerativeModel('gemini-pro-vision')
```

```python
response = model.generate_content(img)

to_markdown(response.text)
```

```python
message_txt = "You are an astrophysicist. This is a Jupiter image and you must locate the plasma waves on the surface of Jupiter, show by the red arrow. Estimate the height in meters for the amplitude of the plasma wave, note the diameter of the jupiter is about 140,000 km. Additionally, identify the plasma layers involved in the waves and specify the direction of the wavefront. Explain how did you get the amplitude of the plasma wave with math equations. "
```

```python
response2 = model.generate_content([message_txt, img], stream=True)
response2.resolve()
to_markdown(response2.text)
```
